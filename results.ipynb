```python
# flip_accuracy_table.ipynb

# Cell 1: Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cell 2: Load CSV
# Replace 'your_file.csv' with your actual filename
df = pd.read_csv('hardware_export_fixed/tall_parameter_sweep_cmax32.csv')

# Cell 3: Preview
df.head()

# Cell 4: Pivot the data
pivot_df = df.pivot(index='flip_p', columns='num_iter', values='accuracy')
pivot_df = pivot_df.sort_index().sort_index(axis=1)

# Cell 5: Show as table
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_df, annot=True, fmt=".2f", cmap="viridis", cbar_kws={'label': 'Accuracy'})
plt.title("Accuracy Table (flip_p vs num_iter)")
plt.xlabel("num_iter")
plt.ylabel("flip_p")
plt.tight_layout()
plt.show()

# Cell 6: Optional  Show best point
best_row = df[df['is_best'] == True]
if not best_row.empty:
    best = best_row.iloc[0]
    print(f"Best configuration  flip_p: {best.flip_p}, num_iter: {best.num_iter}, accuracy: {best.accuracy}%")
```
